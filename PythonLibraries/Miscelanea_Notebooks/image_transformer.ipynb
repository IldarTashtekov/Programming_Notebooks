```python
path_compressed_file ="/content/drive/MyDrive/Colab_Notebooks/Miscelanea_Notebooks/YDRAY-ia.rar"

path_extraction = "/content/zipextract"
password = None

format="rar"

if format=="zip":
  zip_archive = zipfile.ZipFile(path_compressed_file, "r")
  zip_archive.extractall(pwd=password, path=path_extraction)

elif format=="rar":
  Archive(path_compressed_file).extractall(path_extraction)

else:
  print("no more formats suported")


```

```python

path ="/content/zipextract/ia"
output_path='/content/output'

for filename in glob.glob(os.path.join(path, '*.jfif')):    
  
  img = Image.open(filename)
  img_path=f'{output_path}/{filename[23:-5]}.png'  
  print(img_path)
  img.save(img_path)

```
