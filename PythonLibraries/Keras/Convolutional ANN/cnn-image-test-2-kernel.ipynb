```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img = io.imread('../input/cnn-data-sources/Mara.jpeg')
img_gray = color.rgb2gray(img)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
c_v = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1],])


c_h = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1],])


img_h = nd.convolve(img_gray, c_h)
img_v = nd.convolve(img_gray, c_v)
```

```python
cmap = plt.cm.gray

img_v = img_v + img_v.min()*-1
img_v = img_v / img_v.max()

print(img_v.min())
print(img_v.max())


for x in range(len(img)):
    for y in range(len(img[x])):
        pixel = img_v[x][y]
        pixel_neg = 1.0 - img_v[x][y]
        
        if pixel_neg  < 0.6:
            pixel_neg = 0
        
        if pixel < 0.6:
            pixel = 0
        
        img_v[x][y] = pixel + pixel_neg


plt.imshow(img_v, cmap)
```
