```python

#%% [code]
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
from random import random
from random import seed
from math import exp

seed(1)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x, y):
  return np.sin(x*0.4) + np.sin(y*0.4)

res = 100

X = np.linspace(-10, 10, res)
Y = np.linspace(-10, 10, res)

X, Y = np.meshgrid(X, Y)

Z = f(X, Y)

surf = ax.plot_surface(X,Y,Z, cmap=cm.cool)

from numpy.lib.function_base import gradient
level_map = np.linspace(np.min(Z), np.max(Z),res)
plt.figure(figsize=(10,7)) 
plt.contourf(X, Y, Z, levels=level_map,cmap=cm.cool)
plt.colorbar()
plt.title('Descenso del gradiente')


params = np.random.rand(2) * 8 - 4

def f_partial_x_derivative(x,y):
  return np.cos(x*0.4)

def f_partial_y_derivative(x,y):
  return np.cos(y*0.4)

def get_gradient(x, y):
  grad = np.zeros(2)
  grad[0] = f_partial_x_derivative(x,y)
  grad[1] = f_partial_y_derivative(x,y)

  return grad

lr = 0.01

for i in range(1000):
  grad = get_gradient(params[0], params[1])
  for param_i in range(len(params)):
    params[param_i] -= lr * grad[param_i]

  if i % 10 == 0:
    plt.plot(params[0], params[1], 'o', c='r')




```
