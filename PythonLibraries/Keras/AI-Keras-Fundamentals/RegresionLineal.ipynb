```python
import numpy as np
import matplotlib.pyplot as plt

#formula lienal y=b0+b1*x, y=mx*b

#minimos cuadrados  ( (x - media de x) (y - media de y ) ) / (x - media de x)^2 


def estimate_b0_b1(x,y):
    #contamos tamaño de x
    n=np.size(x)

    #calculo de promedios de X y de Y
    m_x=np.mean(x)
    m_y=np.mean(y)

    #calcular sumatoria de XY y de XX
    sumatoria_xy=np.sum((x-m_x)*(y-m_y))
    sumatoria_xx=np.sum(x*(x-m_x))

    #cooficiente de regresion
    b_1=sumatoria_xy/sumatoria_xx
    b_0=m_y - b_1 * m_x

    return(b_0, b_1)


#funcion de graficado
def plot_regression(x,y,b):
    plt.scatter(x,y,color="g",marker="o",s=30)

    #formula lienal y=b0+b1*x
    y_pred=b[0]+b[1]*x

    #graficamos
    plt.plot(x,y_pred,color="b")

    #etiquetado
    plt.xlabel("x_Independiente")
    plt.ylabel("y_Dependiente")
    plt.show()

  #codigo main
def main():
    x=np.array([1,2,3,4,5])
    y=np.array([2,4,3,6,5])

    #obtenemos b1 y b0
    b=estimate_b0_b1(x,y)
    print("Los valores de b0 ={}".format(b[0],b[1]))

    plot_regression(x,y,b)


if __name__=="__main__":
      main()
```
