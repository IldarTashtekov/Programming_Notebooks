```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_gaussian_quantiles
```

```python
#creamos set de datos
#1000 ejemplos
N=1000
gaussian_quantiles=make_gaussian_quantiles( mean=None,
                                            cov=0.1, #distribucion de datos
                                            n_samples=N,
                                            n_features=2, #dos dimensioens
                                            n_classes=2,  #dos clases
                                            shuffle=True,
                                            random_state=None
                                            )


```

```python
X,Y=gaussian_quantiles
Y= Y[:,np.newaxis] 

#nuestra red neuronal tendra descubrir donde se ubica cada clase de las dos que hay
plt.scatter(X[:,0],X[:,1], c=Y[:,0], s=40 ,cmap=plt.cm.Spectral)
```

```python
#activation functions
def sigmoid(x, derivate = False):
    if derivate:
        return np.exp(-x)/(( np.exp(-x) +1)**2)
    else:    
        return 1 / (1 + np.exp(-x))

def relu(x, derivate = False):
    if derivate:
        x[x <= 0] = 0
        x[x > 0 ] = 1
        return x
    else:    
        return np.maximum(0, x)
```

```python
#cost function
def mse(y, y_hat, derivate = False):
  if derivate:
    return (y_hat - y)
  else:
    return np.mean((y_hat - y)**2)
```

```python
#una lista con los parametros de la red
def initialize_params_deep(layers_dim):
  parameters={}
  L=len(layers_dim)
  #en este caso cada W representa cada capa que posee distintos w
  for l in range(0,L-1):
     parameters['W'+str(l+1)]=(np.random.rand(layers_dim[l],layers_dim[+1])*2)-1
     parameters['b'+str(l+1)]=(np.random.rand(1,layers_dim[+1])*2)-1

  return parameters   
```

```python
def train(X_data,lr,params,training=True):
## Forward

    params['A0'] = X_data
        
    params['Z1'] = np.matmul(params['A0'],params['W1']) + params['b1']
    params['A1'] = relu(params['Z1'])
    
    params['Z2'] = np.matmul(params['A1'],params['W2']) + params['b2']
    params['A2'] = relu(params['Z2'])
       
    params['Z3'] = np.matmul(params['A2'],params['W3']) + params['b3']
    params['A3'] = sigmoid(params['Z3'])
  
    output = params['A3']
    
    if training:
    # Backpropagation
    
        params['dZ3'] =  mse(Y,output,True) * sigmoid(params['A3'],True)
        params['dW3'] = np.matmul(params['A2'].T,params['dZ3'])
        
        params['dZ2'] = np.matmul(params['dZ3'],params['W3'].T) * relu(params['A2'],True)
        params['dW2'] = np.matmul(params['A1'].T,params['dZ2'])
        
        params['dZ1'] = np.matmul(params['dZ2'],params['W2'].T) * relu(params['A1'],True)
        params['dW1'] = np.matmul(params['A0'].T,params['dZ1'])

        
        ## Gradinet Descent:
           
        params['W3'] = params['W3'] - params['dW3'] * lr
        params['b3'] = params['b3'] - (np.mean(params['dZ3'],axis=0, keepdims=True)) * lr
        
        params['W2'] = params['W2'] - params['dW2'] * lr
        params['b2'] = params['b2'] - (np.mean(params['dZ2'],axis=0, keepdims=True)) * lr
        
        params['W1'] = params['W1'] -params['dW1'] * lr
        params['b1'] = params['b1'] - (np.mean(params['dZ1'],axis=0, keepdims=True)) * lr
    
    return output
```

```python
layer_dims=[2,4,8,1]
params=initialize_params_deep(layer_dims)

#reidmensionamos X shape
params['W1'].shape
X.shape
print(params['W1'].shape)
print(X.shape)

np.matmul(X,params['W1'])

print(params['W1'].shape)


errors = []
for _ in range(50000):
    output = train(X,0.001,params)
    if _ % 25 == 0:
        print(mse(Y,output))
        errors.append(mse(Y,output))
```
