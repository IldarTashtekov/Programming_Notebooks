###Data wrangling, Domado de datos 🤠

```python
#Añadir columna newspaper id donde todas las filas son 'el universal'
el_universal['newspaper_id'] = 'el universal'

#obtener el host 
el_universal['host'] = el_universal['url'].apply(lambda url : urlparse(url).netloc) #por cada url nos devuelve el host con la funcion urlparse().netloc

el_universal['host'].value_counts() #describrimos que el universal esta compuesto de varios sitios web
```

####datos faltantes

```python
#rellenar datos faltantes

#extraemos las filas con titles vacios
missing_titles_mask = el_universal['title'].isna()
#en las filas donde hay NaNs metemos una columna con el titulo sacado de la url
missing_titles= (el_universal[missing_titles_mask]['url']
                  .str.extract(r'(?P<missing_titles>[^/]+)$') #saca los titulos del final de la url
                  .applymap(lambda title: title.replace('-',' ').capitalize())   #es lo de abajo pero mas legible


                   #.applymap(lambda title: title.split('-'))   # divide el title por cada guion
                  #.applymap(lambda title_word_list: ' '.join(title_word_list)) #juntamos la lista de palabras con un espacio entre ellas
                  )

missing_titles
```

####limpiamos datos

```python
#limpiamos los '\n' que hay al principio del body

#primer paso : añadir uid a las filas, en vd no aporta nada , nose pk lo hicimos
uids = (el_universal
          .apply(lambda row: hashlib.md5(bytes(row['url'].encode('utf-8'))) , axis=1)
          .apply(lambda hash_obj: hash_obj.hexdigest())
        )


el_universal['uid'] = uids  
el_universal.set_index('uid', inplace=True)

el_universal
```

```python
#quitamos los '\n
striped_body = (el_universal
                  .apply(lambda row: row['body'], axis=1) # a cada body
                  .apply(lambda body: list(body)) #transformamos string a list de chars
                  .apply(lambda letters: list(map(lambda letter: letter.replace('\n',''), letters))) #reemplazamos \n por vacio
                  .apply(lambda letters: ''.join(letters))       #jutnamos el string
        
)
                  
striped_body
```

```python
# una forma mil veces mas corta de hacerlo es la siguiente
a = el_universal.apply(lambda row: row['body'].replace('\n', ''), axis=1)
a
```

```python
# datos duplicados

el_universal['title'].value_counts()
```

```python
#con esto eliminas duplicados de un subset en concreto
el_universal.drop_duplicates(subset=['title'], keep='first', inplace = True)
```

####data enrichment

esta parte es una cosa sin importancia

```python
def tokenize_column(df, column_name):
  return (df
          .dropna()
          .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
          .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
          .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
          .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list )))
          .apply(lambda valid_word_list: len(valid_word_list))
          )
```

```python
stop_words = set(stopwords.words('spanish'))

el_universal['n_tokens_title'] = tokenize_column(el_universal, 'title')

el_universal['n_tokens_title']
```
