Secuencias de eventos tipo Bernoulli

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$



```python


# definiciÃ³n de la distribuciÃ³n binomial 
def my_binomial(k, n, p):
#                   (n!/k!               (n-k)!)      p^k         (1-p)^(n-k)    
  return factorial(n)/(factorial(k)*(factorial(n-k)))*pow(p,k)*pow(1-p, n-k)

#mi funcion
my_binomial(2, 3, 0.5)

# es mejor usar la funcion binom(num intentos, probabilidad p) de la libreria scipy
dist = binom(3, 0.5)

dist.pmf(2) #probability mass function (densidad de probabilidad)
```

#####Distribucion normal o gaussiana


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribuciÃ³n
* $\sigma$: desviaciÃ³n estÃ¡ndar de la distribuciÃ³n


```python
# definimos nuestra distribuciÃ³n gaussiana
def gaussian(x, mu, sigma):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))

#el metodo con funciones
x = np.arange(-4,4,0.1)
y = gaussian(x, 0.0, 1.0)

plt.plot(x, y)
```

```python
# usando scipy norm(mu,sigma)
#es muy comun usar media como mu y std como sigma
dist = norm(0, 1)

x = np.arange(-4,4,0.1)
y = [dist.pdf(value) for value in x]

plt.plot(x, y)
```

#####estimacion parametrica

esta basada en forzar los parametros de la distribucion ajustando los datos a la distribucion normal de forma mas exacta

```python
#generamos muestras de datos
#tendra una distribucion normal
sample = np.random.normal(loc = 50, scale = 5, size = 1000) #Donde loc es la media y scale el desvÃ­o estandar

mu = sample.mean()
sigma = sample.std()

#este genera una distribucion normal mas arquetipica
dist = norm(mu,sigma)

#generamos otro set de datos con media 50
values=[value for value in range(30,70)]

#la dsitribucion deberia ser casi igual
probabilidades = [dist.pdf(value) for value in values]

plt.hist(sample, bins=30, density=True)
plt.plot(values, probabilidades)
plt.show()
```

#####estimacion parametrica

CombinaciÃ³n de varias distribuciones para ajustar la distribucion

```python
#generamos una distribucion bimodal jutnando dos distribuciones normales
sample1 = np.random.normal(loc=20, scale=5, size=300)
sample2 = np.random.normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))

#modelo de estimacion, bandwith es el suavizado y kernel funciones de distribuciÃ³n que se usan para la estimaciÃ³n
model = KernelDensity(bandwidth = 2, kernel = 'gaussian')
sample = sample.reshape((len(sample), 1))
model.fit(sample)

#generamos los datos del modelo 
values = np.asarray([value for value in range(1,60)])
values = values.reshape((len(values),1))

#Una aclaraciÃ³n, el cÃ¡lculo de forma logarÃ­timica tiene que ver con la complejidad algorÃ­tmica, es una cuestiÃ³n 
#de eficiencia. Luego a travÃ©s de funciÃ³n inversa, la exponencial, obtenemos las probabilidades.
probabilities = model.score_samples(values) 
probabilities = np.exp(probabilities)

plt.hist(sample, bins=50, density=True) 
plt.plot(values[:], probabilities)
plt.show()
```

##MLE (Maximum Likelihood Estimation)

$$
\max {\sum_{i} log P(Xi, \theta)}
$$


estimacion de maxima verisimilitud

El MLE consiste en ajustar a un conjunto de datos, una distribucion que describa lo mejor posible esos datos.




#####Regresion logistica

es usado para problemas de clasificacion, por ejemplo la funcion sigmoide
la funcion sigmoide separa los datos lienales entre unos mas cercanos al 1 y otros mas cercanos al 0, tenemos que ajustar los pesos de la funcion lienal para que al pasarla por la sigmoide los datos esten mas cerca del 0 o del 1, usando una funcion del coste prediccion, por ejemplo: \
prediccion = 0.9, max = 1

error = prediccion - max = 0.1

<br><br>

#####como usamos el MLE en este tipo de problemas ? 
\

funcion sigmoide le pasamos la x lineal y esa la modifica:
$$
{\hat{y}={\frac {1}{1+e^{-x}}}}
$$

el maximo de la suma de las probabilidades de obtener una y dados una x & una hipotesis de modelamiento theta
$$
\max {\sum_{i} log P(y_i | x_i, \theta)}
$$

los valores reales que debo obtener del data set es $${y}$$ pero los valores que nos arroja la funcion logistica son $${\hat{y}}$$ 

asi que nesesito una verisimilitud que sea maxima con las dos y. \
eso se hace usando bernuli 

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$

que nos da la probabilidad si y=1 y 1-P si y=0

$$
P=\begin{cases} P &  y=1 \\
1-P & y=0 \end{cases} 
$$

<br><br>

$$
\overline{P} = PÂ·1 + (1-P)Â·0
$$

<br><br>
esta funcion toma valores maximos cuando las dos y se parecen <br><br>
$$
L =\hat{y}Â·y  + (1-\hat{y})Â·(1-y)
$$

ahora hay dos puntos que revisar

1- en vez de usar P usaremos log(P) para evitar el underflow
$$
P {\Rightarrow}log(P)
$$
2- sumaremos sobre todas las L para conseguir la L total
$$
{\sum_{}Li = L}
$$

el resultado de la funcion de verosimilirud nos quedaria asi

$$
 L=\sum_{i} Li ( y_i Â· log(\hat{y_i})+ (1-\hat{y_i}) Â· log(\hat{y_i}) )
$$

ahora tocaria maximizar la L, pero si nos fijamos maximizar la L es tmb disminuir la -L

max L = min (-L)

esa funcion se llama cross entropy y es una funcion de costo y hallar el minimo de esta funcion es igual a hallar el maximo de la funcion L anterior

$$
 CE=-\sum_{i} Li ( y_i Â· log(\hat{y_i})+ (1-\hat{y_i}) Â· log(\hat{y_i}) )
$$




```python
def likelihood(y, yp):
  return yp*y+(1-yp)*(1-y)

Y = np.arange(0, 1, 0.01)
YP = np.arange(0, 1, 0.01)
Y, YP = np.meshgrid(Y, YP)
Z = likelihood(Y, YP)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(YP, Y, Z, 50, cmap='binary')
```
